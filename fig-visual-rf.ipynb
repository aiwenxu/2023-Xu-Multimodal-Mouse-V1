```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.filters import butterworth
```

```python
# mouse = "070921_J553RT"
# mouse = "110421_J569LT"
mouse = "101521_J559NC"
```

```python
path_beh = f"gradient_ascent_cnn_gru_shifter/{mouse}"
```

```python
corr_beh = np.load(f"corr_cnn_gru_shifter/{mouse}_seq_1.npy")
n_neur = corr_beh.shape[-1]
idx_sorted = corr_beh[0].argsort()[::-1]
corr_beh_sorted = corr_beh[0][idx_sorted]
```

```python
vmax = {
    '070921_J553RT': .2,
    '110421_J569LT': .15,
    '101521_J559NC': .15
}
```

```python
mouse
```

```python
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(6.5, 12))
for idx, ax in zip(idx_sorted[:32], axes.ravel()):
    vis = np.load(f"{path_beh}/img_{idx}.npy")
    ax.imshow(butterworth(vis, cutoff_frequency_ratio=.05, high_pass=False), cmap='coolwarm', 
              vmin=-vmax[mouse], vmax=vmax[mouse])
    ax.set_xticks([])
    ax.set_yticks([])
    ax.text(35, 55, f"$cc$={corr_beh[0][idx]:.3f}")
fig.tight_layout()
fig.savefig(f"fig-visual-rf-{mouse}.pdf", transparent=True)
```
