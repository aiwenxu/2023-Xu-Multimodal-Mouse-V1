```python
mouse = "070921_J553RT"
corr_vis = np.load("corr_cnn_shifter/" + f"{mouse}_shifter_True_seq_1.npy")
corr_beh = np.load("corr_cnn_gru_shifter/" + f"{mouse}_seq_1.npy")

fig, ax = plt.subplots()
ax.scatter(corr_vis[0], corr_beh[0], alpha=.6, s=100)
ax.plot([0, 1], [0, 1], 'k--')
ax.axis('equal')
fig.savefig(f'scatter-{mouse}.pdf', transparent=True)
```

```python
plt.hist(corr_beh[0] - corr_vis[0], range=(-0.2, 0.2), ec='k', bins=21)
plt.savefig(f'hist-{mouse}.pdf')
```
